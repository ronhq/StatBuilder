```python
# 🏀 Rebounding Load % — 2024-25 Regular Season
# ------------------------------------------------------------
# Dataset columns used:
# PLAYER, TEAM, Year, Season Type, GP, MIN, OREB, DREB, REB
# ------------------------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_rows", 50)
pd.set_option("display.max_columns", None)

# ================================
# 1) Load CSV (Colab or Local)
# ================================
try:
    from google.colab import files
    uploaded = files.upload()
    file_path = list(uploaded.keys())[0]
except:
    file_path = "nba_traditional_stats_2024_2025.csv"  # <-- change if local

print("Using file:", file_path)
df = pd.read_csv(file_path)
df.columns = [c.strip() for c in df.columns]

# Clean Season Type
if "Season Type" in df.columns:
    df["Season Type"] = df["Season Type"].astype(str).str.replace("%20", " ", regex=False)

# Ensure numeric
num_cols = ["GP","MIN","OREB","DREB","REB"]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# ================================
# 2) Filter to 2024-25 Regular Season
# ================================
df_use = df[(df["Year"] == "2024-25") &
            (df["Season Type"].str.lower() == "regular season")].copy()

MIN_GP = 20
MIN_MINUTES = 300
df_use = df_use[(df_use["GP"] >= MIN_GP) & (df_use["MIN"] >= MIN_MINUTES)].copy()

print(f"Filtered to {len(df_use)} players for 2024-25 Regular Season")

# ================================
# 3) Compute Rebounding Load %
# ================================
# Team totals (Year + Season Type + TEAM)
team_reb = (df_use.groupby(["Year","Season Type","TEAM"], as_index=False)
            .agg(TeamREB=("REB","sum"),
                 TeamOREB=("OREB","sum"),
                 TeamDREB=("DREB","sum")))

# Merge back
df_use = df_use.merge(team_reb, on=["Year","Season Type","TEAM"], how="left")

# Individual share of team totals
df_use["REB_Load"] = np.where(df_use["TeamREB"] > 0, df_use["REB"] / df_use["TeamREB"], np.nan) * 100
df_use["OREB_Load"] = np.where(df_use["TeamOREB"] > 0, df_use["OREB"] / df_use["TeamOREB"], np.nan) * 100
df_use["DREB_Load"] = np.where(df_use["TeamDREB"] > 0, df_use["DREB"] / df_use["TeamDREB"], np.nan) * 100

cols_out = ["PLAYER","TEAM","GP","MIN","OREB","DREB","REB",
            "OREB_Load","DREB_Load","REB_Load"]
rebload_tbl = (df_use[cols_out]
               .sort_values("REB_Load", ascending=False)
               .reset_index(drop=True))

print("Sample of Rebounding Load table:")
display(rebload_tbl.head(10))

# ================================
# 4) Leaderboards
# ================================
TOPN = 15
leaders_total = rebload_tbl.head(TOPN)
leaders_off = rebload_tbl.sort_values("OREB_Load", ascending=False).head(TOPN)
leaders_def = rebload_tbl.sort_values("DREB_Load", ascending=False).head(TOPN)

print("\nTop by Total Rebounding Load %:")
display(leaders_total)

print("\nTop by Offensive Rebounding Load %:")
display(leaders_off)

print("\nTop by Defensive Rebounding Load %:")
display(leaders_def)

# ================================
# 5) Plots
# ================================
def barh_plot(df_plot, col, title, xlabel, fname, color="darkcyan"):
    plt.figure(figsize=(10, 7))
    plt.barh(df_plot["PLAYER"][::-1], df_plot[col][::-1], color=color)
    for i, v in enumerate(df_plot[col][::-1]):
        plt.text(v, i, f"{v:.1f}%", va="center", ha="left")
    plt.xlabel(xlabel)
    plt.title(title)
    plt.tight_layout()
    plt.savefig(fname, dpi=300)
    plt.show()

barh_plot(leaders_total, "REB_Load",
          f"Top {TOPN} Rebounding Load % — 2024-25 RS",
          "Share of Team Rebounds (%)",
          "rebounding_load_total.png")

barh_plot(leaders_off, "OREB_Load",
          f"Top {TOPN} Offensive Rebounding Load % — 2024-25 RS",
          "Share of Team Offensive Rebounds (%)",
          "rebounding_load_off.png")

barh_plot(leaders_def, "DREB_Load",
          f"Top {TOPN} Defensive Rebounding Load % — 2024-25 RS",
          "Share of Team Defensive Rebounds (%)",
          "rebounding_load_def.png")

# ================================
# 6) Save Outputs
# ================================
rebload_tbl.to_csv("rebounding_load_2024_25.csv", index=False)
leaders_total.to_csv("leaders_rebounding_load_total_2024_25.csv", index=False)
leaders_off.to_csv("leaders_rebounding_load_off_2024_25.csv", index=False)
leaders_def.to_csv("leaders_rebounding_load_def_2024_25.csv", index=False)

print("\nSaved:")
print("- rebounding_load_2024_25.csv (full table)")
print("- leaders_rebounding_load_total_2024_25.csv (total load)")
print("- leaders_rebounding_load_off_2024_25.csv (offensive load)")
print("- leaders_rebounding_load_def_2024_25.csv (defensive load)")
print("- rebounding_load_total.png")
print("- rebounding_load_off.png")
print("- rebounding_load_def.png")

# ================================
# 7) Zip & Download (Colab)
# ================================
import os, zipfile
from google.colab import files

outputs = [
    "rebounding_load_2024_25.csv",
    "leaders_rebounding_load_total_2024_25.csv",
    "leaders_rebounding_load_off_2024_25.csv",
    "leaders_rebounding_load_def_2024_25.csv",
    "rebounding_load_total.png",
    "rebounding_load_off.png",
    "rebounding_load_def.png"
]

zip_name = "rebounding_load_outputs_2024_25.zip"

with zipfile.ZipFile(zip_name, "w") as z:
    for f in outputs:
        if os.path.exists(f):
            z.write(f)

files.download(zip_name)
```
