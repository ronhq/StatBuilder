```python
# 🏀 EFF Leaders — 2024-25 Regular Season
# ------------------------------------------------------------
# Uses ONLY your scraped columns:
# PLAYER, TEAM, Year, Season Type, GP, MIN,
# FGM, FGA, FTM, FTA, PTS, REB, AST, STL, BLK, TOV, EFF (optional)
# ------------------------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_rows", 50)
pd.set_option("display.max_columns", None)

# ================================
# 1) Load CSV (Colab or Local)
# ================================
try:
    from google.colab import files
    uploaded = files.upload()
    file_path = list(uploaded.keys())[0]
except:
    file_path = "nba_traditional_stats_2024_2025.csv"   # <-- change if local

print("Using file:", file_path)
df = pd.read_csv(file_path)
df.columns = [c.strip() for c in df.columns]

# Clean Season Type text
if "Season Type" in df.columns:
    df["Season Type"] = df["Season Type"].astype(str).str.replace("%20", " ", regex=False)

# Ensure numeric
num_cols = ["GP","MIN","FGM","FGA","FTM","FTA","PTS","REB","AST","STL","BLK","TOV","EFF"]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# ================================
# 2) Filter to 2024-25 Regular Season + thresholds
# ================================
df_use = df[(df["Year"] == "2024-25") &
            (df["Season Type"].str.lower() == "regular season")].copy()

# Minimum samples to avoid noise
MIN_GP = 20
MIN_MINUTES = 300

df_use = df_use[(df_use["GP"] >= MIN_GP) &
                (df_use["MIN"] >= MIN_MINUTES)].copy()

print(f"Qualified players: {len(df_use)}")

# ================================
# 3) Compute EFF Variants
# ================================
# Totals-based EFF (classic box EFF on season totals)
# EFF_total = (PTS + REB + AST + STL + BLK) - (FGA - FGM) - (FTA - FTM) - TOV
df_use["EFF_total_calc"] = (
    df_use.get("PTS", 0) + df_use.get("REB", 0) + df_use.get("AST", 0) +
    df_use.get("STL", 0) + df_use.get("BLK", 0)
    - (df_use.get("FGA", 0) - df_use.get("FGM", 0))
    - (df_use.get("FTA", 0) - df_use.get("FTM", 0))
    - df_use.get("TOV", 0)
)

# Per-game variant
df_use["PPG"] = np.where(df_use["GP"] > 0, df_use["PTS"] / df_use["GP"], np.nan)
df_use["RPG"] = np.where(df_use["GP"] > 0, df_use["REB"] / df_use["GP"], np.nan)
df_use["APG"] = np.where(df_use["GP"] > 0, df_use["AST"] / df_use["GP"], np.nan)
df_use["SPG"] = np.where(df_use["GP"] > 0, df_use["STL"] / df_use["GP"], np.nan)
df_use["BPG"] = np.where(df_use["GP"] > 0, df_use["BLK"] / df_use["GP"], np.nan)
df_use["FGA_pg"] = np.where(df_use["GP"] > 0, df_use["FGA"] / df_use["GP"], np.nan)
df_use["FGM_pg"] = np.where(df_use["GP"] > 0, df_use["FGM"] / df_use["GP"], np.nan)
df_use["FTA_pg"] = np.where(df_use["GP"] > 0, df_use["FTA"] / df_use["GP"], np.nan)
df_use["FTM_pg"] = np.where(df_use["GP"] > 0, df_use["FTM"] / df_use["GP"], np.nan)
df_use["TOV_pg"] = np.where(df_use["GP"] > 0, df_use["TOV"] / df_use["GP"], np.nan)

df_use["EFF_per_game_calc"] = (
    df_use["PPG"] + df_use["RPG"] + df_use["APG"] + df_use["SPG"] + df_use["BPG"]
    - (df_use["FGA_pg"] - df_use["FGM_pg"])
    - (df_use["FTA_pg"] - df_use["FTM_pg"])
    - df_use["TOV_pg"]
)

# Per-36 variant (normalizes for playing time)
df_use["PTS_per36"] = df_use["PTS"] / df_use["MIN"] * 36
df_use["REB_per36"] = df_use["REB"] / df_use["MIN"] * 36
df_use["AST_per36"] = df_use["AST"] / df_use["MIN"] * 36
df_use["STL_per36"] = df_use["STL"] / df_use["MIN"] * 36
df_use["BLK_per36"] = df_use["BLK"] / df_use["MIN"] * 36
df_use["FGA_per36"] = df_use["FGA"] / df_use["MIN"] * 36
df_use["FGM_per36"] = df_use["FGM"] / df_use["MIN"] * 36
df_use["FTA_per36"] = df_use["FTA"] / df_use["MIN"] * 36
df_use["FTM_per36"] = df_use["FTM"] / df_use["MIN"] * 36
df_use["TOV_per36"] = df_use["TOV"] / df_use["MIN"] * 36

df_use["EFF_per36_calc"] = (
    df_use["PTS_per36"] + df_use["REB_per36"] + df_use["AST_per36"] +
    df_use["STL_per36"] + df_use["BLK_per36"]
    - (df_use["FGA_per36"] - df_use["FGM_per36"])
    - (df_use["FTA_per36"] - df_use["FTM_per36"])
    - df_use["TOV_per36"]
)

# Keep scraped EFF if it exists (often per-game in some sources)
has_eff = "EFF" in df_use.columns

# ================================
# 4) Leaderboards
# ================================
TOPN = 20

leaders_total = (df_use.sort_values("EFF_total_calc", ascending=False)
                 .loc[:, ["PLAYER","TEAM","GP","MIN","EFF_total_calc"]]
                 .head(TOPN))

leaders_pg = (df_use.sort_values("EFF_per_game_calc", ascending=False)
              .loc[:, ["PLAYER","TEAM","GP","MIN","EFF_per_game_calc"]]
              .head(TOPN))

leaders_p36 = (df_use.sort_values("EFF_per36_calc", ascending=False)
               .loc[:, ["PLAYER","TEAM","GP","MIN","EFF_per36_calc"]]
               .head(TOPN))

if has_eff:
    leaders_scraped = (df_use.sort_values("EFF", ascending=False)
                       .loc[:, ["PLAYER","TEAM","GP","MIN","EFF"]]
                       .head(TOPN))

print("\nTop by EFF (Totals-based):")
display(leaders_total)

print("\nTop by EFF per game (calculated):")
display(leaders_pg)

print("\nTop by EFF per 36 (calculated):")
display(leaders_p36)

if has_eff:
    print("\nTop by EFF (scraped column):")
    display(leaders_scraped)

# ================================
# 5) Plots
# ================================
def barh(df_plot, col, title, xlabel, fname):
    plt.figure(figsize=(10, 7))
    plt.barh(df_plot["PLAYER"][::-1], df_plot[col][::-1])
    for i, v in enumerate(df_plot[col][::-1]):
        plt.text(v, i, f"{v:.1f}", va="center", ha="left")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.tight_layout()
    plt.savefig(fname, dpi=300)
    plt.show()

barh(leaders_total, "EFF_total_calc",
     f"Top {TOPN} EFF Leaders (Totals) — 2024-25 RS", "EFF (totals)",
     "eff_leaders_totals.png")

barh(leaders_pg, "EFF_per_game_calc",
     f"Top {TOPN} EFF Leaders (Per Game) — 2024-25 RS", "EFF per game (calc)",
     "eff_leaders_per_game.png")

barh(leaders_p36, "EFF_per36_calc",
     f"Top {TOPN} EFF Leaders (Per 36) — 2024-25 RS", "EFF per 36 (calc)",
     "eff_leaders_per36.png")

if has_eff:
    barh(leaders_scraped, "EFF",
         f"Top {TOPN} EFF Leaders (Scraped) — 2024-25 RS", "EFF (scraped)",
         "eff_leaders_scraped.png")

# ================================
# 6) Save CSVs
# ================================
# Full table export
cols_export = ["PLAYER","TEAM","GP","MIN",
               "EFF_total_calc","EFF_per_game_calc","EFF_per36_calc"]
if has_eff:
    cols_export.append("EFF")

eff_table = df_use[cols_export].sort_values("EFF_per_game_calc", ascending=False)
eff_table.to_csv("eff_leaders_full_2024_25.csv", index=False)

leaders_total.to_csv("eff_leaders_totals_top20_2024_25.csv", index=False)
leaders_pg.to_csv("eff_leaders_per_game_top20_2024_25.csv", index=False)
leaders_p36.to_csv("eff_leaders_per36_top20_2024_25.csv", index=False)
if has_eff:
    leaders_scraped.to_csv("eff_leaders_scraped_top20_2024_25.csv", index=False)

print("\nSaved:")
print("- eff_leaders_full_2024_25.csv")
print("- eff_leaders_totals_top20_2024_25.csv")
print("- eff_leaders_per_game_top20_2024_25.csv")
print("- eff_leaders_per36_top20_2024_25.csv")
if has_eff:
    print("- eff_leaders_scraped_top20_2024_25.csv")
print("- eff_leaders_totals.png")
print("- eff_leaders_per_game.png")
print("- eff_leaders_per36.png")
if has_eff:
    print("- eff_leaders_scraped.png")

# ================================
# 7) (Optional) Zip & Download (Colab)
# ================================
try:
    import os, zipfile
    from google.colab import files
    bundle = [
        "eff_leaders_full_2024_25.csv",
        "eff_leaders_totals_top20_2024_25.csv",
        "eff_leaders_per_game_top20_2024_25.csv",
        "eff_leaders_per36_top20_2024_25.csv",
        "eff_leaders_totals.png",
        "eff_leaders_per_game.png",
        "eff_leaders_per36.png",
    ]
    if has_eff:
        bundle += ["eff_leaders_scraped_top20_2024_25.csv", "eff_leaders_scraped.png"]

    zip_name = "eff_leaders_outputs_2024_25.zip"
    with zipfile.ZipFile(zip_name, "w") as z:
        for f in bundle:
            if os.path.exists(f):
                z.write(f)
    files.download(zip_name)
except Exception as e:
    print("Zip/Download skipped (likely not in Colab):", e)
```
