```python
# 🏀 All-Around Impact (Stat-Stuffer Index) — 2024-25 Regular Season
# ------------------------------------------------------------
# Dataset columns used:
# PLAYER, TEAM, Year, Season Type, GP, MIN, REB, AST, STL, BLK, PTS
# ------------------------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_rows", 50)
pd.set_option("display.max_columns", None)

# ================================
# 1) Load CSV (Colab or Local)
# ================================
try:
    from google.colab import files
    uploaded = files.upload()
    file_path = list(uploaded.keys())[0]
except:
    file_path = "nba_traditional_stats_2024_2025.csv"   # <-- change if local

print("Using file:", file_path)
df = pd.read_csv(file_path)
df.columns = [c.strip() for c in df.columns]

# Clean Season Type
if "Season Type" in df.columns:
    df["Season Type"] = df["Season Type"].astype(str).str.replace("%20", " ", regex=False)

# Ensure numeric
num_cols = ["GP","MIN","REB","AST","STL","BLK","PTS"]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# ================================
# 2) Filter to 2024-25 Regular Season
# ================================
df_use = df[(df["Year"] == "2024-25") &
            (df["Season Type"].str.lower() == "regular season")].copy()

MIN_GP = 20
MIN_MINUTES = 300
df_use = df_use[(df_use["GP"] >= MIN_GP) & (df_use["MIN"] >= MIN_MINUTES)].copy()

print(f"Filtered to {len(df_use)} players for 2024-25 Regular Season")

# ================================
# 3) Compute Stat-Stuffer Index
# ================================
# Per-game values
df_use["REB_pg"] = df_use["REB"] / df_use["GP"]
df_use["AST_pg"] = df_use["AST"] / df_use["GP"]
df_use["STL_pg"] = df_use["STL"] / df_use["GP"]
df_use["BLK_pg"] = df_use["BLK"] / df_use["GP"]

# Stat-Stuffer Index = REB + AST + STL + BLK per game
df_use["StatStuffer_pg"] = df_use["REB_pg"] + df_use["AST_pg"] + df_use["STL_pg"] + df_use["BLK_pg"]

# Per-36 version
df_use["REB_per36"] = (df_use["REB"] / df_use["MIN"]) * 36
df_use["AST_per36"] = (df_use["AST"] / df_use["MIN"]) * 36
df_use["STL_per36"] = (df_use["STL"] / df_use["MIN"]) * 36
df_use["BLK_per36"] = (df_use["BLK"] / df_use["MIN"]) * 36
df_use["StatStuffer_per36"] = df_use["REB_per36"] + df_use["AST_per36"] + df_use["STL_per36"] + df_use["BLK_per36"]

cols_out = ["PLAYER","TEAM","GP","MIN","PTS",
            "REB_pg","AST_pg","STL_pg","BLK_pg","StatStuffer_pg",
            "REB_per36","AST_per36","STL_per36","BLK_per36","StatStuffer_per36"]
impact_tbl = (df_use[cols_out]
              .sort_values("StatStuffer_per36", ascending=False)
              .reset_index(drop=True))

print("Sample of All-Around Impact (Stat-Stuffer Index):")
display(impact_tbl.head(10))

# ================================
# 4) Leaderboards
# ================================
TOPN = 15
leaders_pg = impact_tbl.sort_values("StatStuffer_pg", ascending=False).head(TOPN)
leaders_per36 = impact_tbl.sort_values("StatStuffer_per36", ascending=False).head(TOPN)

print("\nTop Stat-Stuffers (Per Game):")
display(leaders_pg)

print("\nTop Stat-Stuffers (Per 36 Minutes):")
display(leaders_per36)

# ================================
# 5) Plots
# ================================
def plot_bar(df_plot, col, title, xlabel, fname, color="navy"):
    plt.figure(figsize=(10, 7))
    plt.barh(df_plot["PLAYER"][::-1], df_plot[col][::-1], color=color)
    for i, v in enumerate(df_plot[col][::-1]):
        plt.text(v, i, f"{v:.2f}", va="center", ha="left")
    plt.xlabel(xlabel)
    plt.title(title)
    plt.tight_layout()
    plt.savefig(fname, dpi=300)
    plt.show()

plot_bar(leaders_pg, "StatStuffer_pg",
         f"Top {TOPN} Stat-Stuffers (Per Game) — 2024-25 RS",
         "Stat-Stuffer Index (REB+AST+STL+BLK per game)",
         "stat_stuffers_pg.png")

plot_bar(leaders_per36, "StatStuffer_per36",
         f"Top {TOPN} Stat-Stuffers (Per 36) — 2024-25 RS",
         "Stat-Stuffer Index (REB+AST+STL+BLK per 36)",
         "stat_stuffers_per36.png")

# ================================
# 6) Save Outputs
# ================================
impact_tbl.to_csv("stat_stuffer_index_2024_25.csv", index=False)
leaders_pg.to_csv("leaders_stat_stuffers_pg_2024_25.csv", index=False)
leaders_per36.to_csv("leaders_stat_stuffers_per36_2024_25.csv", index=False)

print("\nSaved:")
print("- stat_stuffer_index_2024_25.csv (full table)")
print("- leaders_stat_stuffers_pg_2024_25.csv (per game leaders)")
print("- leaders_stat_stuffers_per36_2024_25.csv (per 36 leaders)")
print("- stat_stuffers_pg.png")
print("- stat_stuffers_per36.png")

# ================================
# 7) Zip & Download (Colab)
# ================================
import os, zipfile
from google.colab import files

outputs = [
    "stat_stuffer_index_2024_25.csv",
    "leaders_stat_stuffers_pg_2024_25.csv",
    "leaders_stat_stuffers_per36_2024_25.csv",
    "stat_stuffers_pg.png",
    "stat_stuffers_per36.png"
]

zip_name = "stat_stuffer_outputs_2024_25.zip"

with zipfile.ZipFile(zip_name, "w") as z:
    for f in outputs:
        if os.path.exists(f):
            z.write(f)

files.download(zip_name)
```
