```python
#  Discipline vs Hustle Index  2024-25 Regular Season
# ------------------------------------------------------------
# Dataset columns used:
# PLAYER, TEAM, Year, Season Type, GP, MIN, STL, BLK, PF
# ------------------------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_rows", 50)
pd.set_option("display.max_columns", None)

# ================================
# 1) Load CSV (Colab or Local)
# ================================
try:
    from google.colab import files
    uploaded = files.upload()
    file_path = list(uploaded.keys())[0]
except:
    file_path = "nba_traditional_stats_2024_2025.csv"   # <-- change if local

print("Using file:", file_path)
df = pd.read_csv(file_path)
df.columns = [c.strip() for c in df.columns]

# Clean Season Type text
if "Season Type" in df.columns:
    df["Season Type"] = df["Season Type"].astype(str).str.replace("%20", " ", regex=False)

# Ensure numeric
num_cols = ["GP","MIN","STL","BLK","PF"]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# ================================
# 2) Filter to 2024-25 Regular Season
# ================================
df_use = df[(df["Year"] == "2024-25") &
            (df["Season Type"].str.lower() == "regular season")].copy()

# Minimum filters
MIN_GP = 20
MIN_MINUTES = 300

df_use = df_use[(df_use["GP"] >= MIN_GP) &
                (df_use["MIN"] >= MIN_MINUTES)].copy()

print(f"Filtered to {len(df_use)} players for 2024-25 Regular Season")

# ================================
# 3) Compute Discipline vs Hustle Index
# ================================
# Per 36 calculations
df_use["STL_per36"] = (df_use["STL"] / df_use["MIN"]) * 36
df_use["BLK_per36"] = (df_use["BLK"] / df_use["MIN"]) * 36
df_use["PF_per36"]  = (df_use["PF"]  / df_use["MIN"]) * 36

# Hustle = STL + BLK per 36
df_use["Hustle_per36"] = df_use["STL_per36"] + df_use["BLK_per36"]

# Discipline vs Hustle Index = (STL+BLK) / PF
df_use["Discipline_Hustle_Index"] = np.where(df_use["PF_per36"] > 0,
                                             df_use["Hustle_per36"] / df_use["PF_per36"],
                                             np.nan)

cols_out = ["PLAYER","TEAM","GP","MIN","STL","BLK","PF",
            "STL_per36","BLK_per36","PF_per36","Hustle_per36","Discipline_Hustle_Index"]

dh_tbl = (df_use[cols_out]
          .sort_values("Discipline_Hustle_Index", ascending=False)
          .reset_index(drop=True))

print("Sample of Discipline vs Hustle Index:")
display(dh_tbl.head(10))

# ================================
# 4) Leaderboards
# ================================
TOPN = 15
leaders_best = dh_tbl.head(TOPN)   # high index = high steals+blocks, low fouls
leaders_worst = dh_tbl.sort_values("Discipline_Hustle_Index", ascending=True).head(TOPN)

print("\nTop Smart Defenders (High Hustle, Low Fouls):")
display(leaders_best)

print("\nMost Reckless Defenders (Low Hustle, High Fouls):")
display(leaders_worst)

# ================================
# 5) Plots
# ================================
# Top disciplined defenders
plt.figure(figsize=(10, 7))
plt.barh(leaders_best["PLAYER"][::-1], leaders_best["Discipline_Hustle_Index"][::-1], color="green")
for i, v in enumerate(leaders_best["Discipline_Hustle_Index"][::-1]):
    plt.text(v, i, f"{v:.2f}", va="center", ha="left")
plt.xlabel("Discipline vs Hustle Index ((STL+BLK)/PF per 36)")
plt.title(f"Top {TOPN} Smart Defenders  2024-25 Regular Season")
plt.tight_layout()
plt.savefig("top_disciplined_defenders.png", dpi=300)
plt.show()

# Most reckless defenders
plt.figure(figsize=(10, 7))
plt.barh(leaders_worst["PLAYER"][::-1], leaders_worst["Discipline_Hustle_Index"][::-1], color="red")
for i, v in enumerate(leaders_worst["Discipline_Hustle_Index"][::-1]):
    plt.text(v, i, f"{v:.2f}", va="center", ha="left")
plt.xlabel("Discipline vs Hustle Index ((STL+BLK)/PF per 36)")
plt.title(f"Top {TOPN} Reckless Defenders  2024-25 Regular Season")
plt.tight_layout()
plt.savefig("top_reckless_defenders.png", dpi=300)
plt.show()

# Scatter: PF per 36 vs Hustle per 36
plt.figure(figsize=(9, 7))
plt.scatter(df_use["PF_per36"], df_use["Hustle_per36"],
            s=np.clip(df_use["MIN"]/10, 20, 400),
            alpha=0.6, c="blue")
plt.xlabel("PF per 36")
plt.ylabel("Hustle per 36 (STL+BLK)")
plt.title("Discipline vs Hustle  2024-25 Regular Season (bubble size ~ minutes)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("scatter_discipline_vs_hustle.png", dpi=300)
plt.show()

# ================================
# 6) Save Outputs
# ================================
dh_tbl.to_csv("discipline_vs_hustle_2024_25.csv", index=False)
leaders_best.to_csv("leaders_disciplined_defenders_2024_25.csv", index=False)
leaders_worst.to_csv("leaders_reckless_defenders_2024_25.csv", index=False)

print("\nSaved:")
print("- discipline_vs_hustle_2024_25.csv (full table)")
print("- leaders_disciplined_defenders_2024_25.csv (top disciplined)")
print("- leaders_reckless_defenders_2024_25.csv (top reckless)")
print("- top_disciplined_defenders.png")
print("- top_reckless_defenders.png")
print("- scatter_discipline_vs_hustle.png")
```
