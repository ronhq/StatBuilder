```python
#  Turnover Economy  2024-25 Regular Season
# ------------------------------------------------------------
# Dataset columns used:
# PLAYER, TEAM, Year, Season Type, GP, MIN, AST, TOV
# ------------------------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_rows", 50)
pd.set_option("display.max_columns", None)

# ================================
# 1) Load CSV (Colab or Local)
# ================================
try:
    from google.colab import files
    uploaded = files.upload()
    file_path = list(uploaded.keys())[0]
except:
    file_path = "nba_traditional_stats_2024_2025.csv"   # <-- change if local

print("Using file:", file_path)
df = pd.read_csv(file_path)
df.columns = [c.strip() for c in df.columns]

# Clean Season Type
if "Season Type" in df.columns:
    df["Season Type"] = df["Season Type"].astype(str).str.replace("%20", " ", regex=False)

# Ensure numeric
num_cols = ["GP","MIN","AST","TOV"]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# ================================
# 2) Filter to 2024-25 Regular Season
# ================================
df_use = df[(df["Year"] == "2024-25") &
            (df["Season Type"].str.lower() == "regular season")].copy()

MIN_GP = 20
MIN_MINUTES = 300
df_use = df_use[(df_use["GP"] >= MIN_GP) & (df_use["MIN"] >= MIN_MINUTES)].copy()

print(f"Filtered to {len(df_use)} players for 2024-25 Regular Season")

# ================================
# 3) Compute Turnover Economy Metrics
# ================================
# TOV per 36 minutes
df_use["TOV_per36"] = (df_use["TOV"] / df_use["MIN"]) * 36

# AST/TOV ratio (already in dataset but recompute for safety)
df_use["AST_TOV_ratio"] = np.where(df_use["TOV"] > 0, df_use["AST"] / df_use["TOV"], np.nan)

cols_out = ["PLAYER","TEAM","GP","MIN","AST","TOV","TOV_per36","AST_TOV_ratio"]
tov_tbl = (df_use[cols_out]
           .sort_values("TOV_per36", ascending=True)  # safest first
           .reset_index(drop=True))

print("Sample Turnover Economy table:")
display(tov_tbl.head(10))

# ================================
# 4) Leaderboards
# ================================
TOPN = 15
# Safest ballhandlers (lowest TOV per 36)
leaders_safe = tov_tbl.head(TOPN)

# Sloppiest (highest TOV per 36)
leaders_sloppy = tov_tbl.sort_values("TOV_per36", ascending=False).head(TOPN)

# Best AST/TOV ratio
leaders_ast_tov = df_use.sort_values("AST_TOV_ratio", ascending=False).head(TOPN)[cols_out]

print("\nTop Safest Ballhandlers (Lowest TOV per 36):")
display(leaders_safe)

print("\nTop Sloppiest Players (Highest TOV per 36):")
display(leaders_sloppy)

print("\nTop by AST/TOV Ratio:")
display(leaders_ast_tov)

# ================================
# 5) Plots
# ================================
# Safest
plt.figure(figsize=(10, 7))
plt.barh(leaders_safe["PLAYER"][::-1], leaders_safe["TOV_per36"][::-1], color="green")
for i, v in enumerate(leaders_safe["TOV_per36"][::-1]):
    plt.text(v, i, f"{v:.2f}", va="center", ha="left")
plt.xlabel("Turnovers per 36")
plt.title(f"Top {TOPN} Safest Ballhandlers  2024-25 RS")
plt.tight_layout()
plt.savefig("tov_safe_leaders.png", dpi=300)
plt.show()

# Sloppiest
plt.figure(figsize=(10, 7))
plt.barh(leaders_sloppy["PLAYER"][::-1], leaders_sloppy["TOV_per36"][::-1], color="red")
for i, v in enumerate(leaders_sloppy["TOV_per36"][::-1]):
    plt.text(v, i, f"{v:.2f}", va="center", ha="left")
plt.xlabel("Turnovers per 36")
plt.title(f"Top {TOPN} Sloppiest Players  2024-25 RS")
plt.tight_layout()
plt.savefig("tov_sloppy_leaders.png", dpi=300)
plt.show()

# AST/TOV ratio
plt.figure(figsize=(10, 7))
plt.barh(leaders_ast_tov["PLAYER"][::-1], leaders_ast_tov["AST_TOV_ratio"][::-1], color="blue")
for i, v in enumerate(leaders_ast_tov["AST_TOV_ratio"][::-1]):
    plt.text(v, i, f"{v:.2f}", va="center", ha="left")
plt.xlabel("AST/TOV Ratio")
plt.title(f"Top {TOPN} AST/TOV Ratio Leaders  2024-25 RS")
plt.tight_layout()
plt.savefig("ast_tov_ratio_leaders.png", dpi=300)
plt.show()

# ================================
# 6) Save Outputs
# ================================
tov_tbl.to_csv("turnover_economy_2024_25.csv", index=False)
leaders_safe.to_csv("leaders_tov_safe_2024_25.csv", index=False)
leaders_sloppy.to_csv("leaders_tov_sloppy_2024_25.csv", index=False)
leaders_ast_tov.to_csv("leaders_ast_tov_ratio_2024_25.csv", index=False)

print("\nSaved:")
print("- turnover_economy_2024_25.csv (full table)")
print("- leaders_tov_safe_2024_25.csv (safest)")
print("- leaders_tov_sloppy_2024_25.csv (sloppiest)")
print("- leaders_ast_tov_ratio_2024_25.csv (best ratios)")
print("- tov_safe_leaders.png")
print("- tov_sloppy_leaders.png")
print("- ast_tov_ratio_leaders.png")

# ================================
# 7) Zip & Download
# ================================
import os, zipfile
from google.colab import files

outputs = [
    "turnover_economy_2024_25.csv",
    "leaders_tov_safe_2024_25.csv",
    "leaders_tov_sloppy_2024_25.csv",
    "leaders_ast_tov_ratio_2024_25.csv",
    "tov_safe_leaders.png",
    "tov_sloppy_leaders.png",
    "ast_tov_ratio_leaders.png"
]

zip_name = "turnover_economy_outputs_2024_25.zip"

with zipfile.ZipFile(zip_name, "w") as z:
    for f in outputs:
        if os.path.exists(f):
            z.write(f)

files.download(zip_name)
```
