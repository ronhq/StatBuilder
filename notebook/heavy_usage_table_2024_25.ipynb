```python
# 🏀 Heavy Usage Proxy (FGA + 0.44*FTA) — 2024-25 Regular Season
# ------------------------------------------------------------
# Dataset columns used:
# PLAYER, TEAM, Year, Season Type, GP, MIN, FGA, FTA, PTS
# ------------------------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_rows", 50)
pd.set_option("display.max_columns", None)

# ================================
# 1) Load CSV (Colab or Local)
# ================================
try:
    from google.colab import files
    uploaded = files.upload()
    file_path = list(uploaded.keys())[0]
except:
    file_path = "nba_traditional_stats_2024_2025.csv"   # <-- change if local

print("Using file:", file_path)
df = pd.read_csv(file_path)
df.columns = [c.strip() for c in df.columns]

# Clean Season Type text
if "Season Type" in df.columns:
    df["Season Type"] = df["Season Type"].astype(str).str.replace("%20", " ", regex=False)

# Ensure numeric
num_cols = ["GP","MIN","FGA","FTA","PTS"]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# ================================
# 2) Filter to 2024-25 Regular Season
# ================================
df_use = df[(df["Year"] == "2024-25") &
            (df["Season Type"].str.lower() == "regular season")].copy()

MIN_GP = 20
MIN_MINUTES = 300
df_use = df_use[(df_use["GP"] >= MIN_GP) & (df_use["MIN"] >= MIN_MINUTES)].copy()

print(f"Filtered to {len(df_use)} players for 2024-25 Regular Season")

# ================================
# 3) Compute Heavy Usage Proxy
# ================================
# Possessions used proxy = FGA + 0.44 * FTA
df_use["Usage_Poss"] = df_use["FGA"] + 0.44 * df_use["FTA"]

# Per game
df_use["Usage_pg"] = np.where(df_use["GP"] > 0, df_use["Usage_Poss"] / df_use["GP"], np.nan)

# Per 36 minutes
df_use["Usage_per36"] = np.where(df_use["MIN"] > 0, df_use["Usage_Poss"] / df_use["MIN"] * 36, np.nan)

# Team share
team_totals = (df_use.groupby(["Year","Season Type","TEAM"], as_index=False)
               .agg(TeamUsage=("Usage_Poss","sum")))
df_use = df_use.merge(team_totals, on=["Year","Season Type","TEAM"], how="left")
df_use["Usage_SharePct"] = np.where(df_use["TeamUsage"] > 0, df_use["Usage_Poss"] / df_use["TeamUsage"] * 100, np.nan)

cols_out = ["PLAYER","TEAM","GP","MIN","PTS","FGA","FTA","Usage_pg","Usage_per36","Usage_SharePct"]
usage_tbl = (df_use[cols_out]
             .sort_values("Usage_pg", ascending=False)
             .reset_index(drop=True))

print("Sample Heavy Usage table:")
display(usage_tbl.head(10))

# ================================
# 4) Leaderboards
# ================================
TOPN = 15
leaders_pg = usage_tbl.sort_values("Usage_pg", ascending=False).head(TOPN)
leaders_p36 = usage_tbl.sort_values("Usage_per36", ascending=False).head(TOPN)
leaders_share = usage_tbl.sort_values("Usage_SharePct", ascending=False).head(TOPN)

print("\nTop by Usage per Game:")
display(leaders_pg)

print("\nTop by Usage per 36:")
display(leaders_p36)

print("\nTop by Team Usage Share %:")
display(leaders_share)

# ================================
# 5) Plots
# ================================
def plot_bar(df_plot, col, title, xlabel, fname, color="purple"):
    plt.figure(figsize=(10, 7))
    plt.barh(df_plot["PLAYER"][::-1], df_plot[col][::-1], color=color)
    for i, v in enumerate(df_plot[col][::-1]):
        plt.text(v, i, f"{v:.1f}", va="center", ha="left")
    plt.xlabel(xlabel)
    plt.title(title)
    plt.tight_layout()
    plt.savefig(fname, dpi=300)
    plt.show()

plot_bar(leaders_pg, "Usage_pg",
         f"Top {TOPN} Heavy Usage (per game) — 2024-25 RS",
         "Usage Possessions per game",
         "heavy_usage_pg.png")

plot_bar(leaders_p36, "Usage_per36",
         f"Top {TOPN} Heavy Usage (per 36) — 2024-25 RS",
         "Usage Possessions per 36",
         "heavy_usage_per36.png")

plot_bar(leaders_share, "Usage_SharePct",
         f"Top {TOPN} Team Usage Share — 2024-25 RS",
         "% of Team Usage",
         "heavy_usage_share.png")

# ================================
# 6) Save Outputs
# ================================
usage_tbl.to_csv("heavy_usage_table_2024_25.csv", index=False)
leaders_pg.to_csv("heavy_usage_pg_leaders_2024_25.csv", index=False)
leaders_p36.to_csv("heavy_usage_per36_leaders_2024_25.csv", index=False)
leaders_share.to_csv("heavy_usage_share_leaders_2024_25.csv", index=False)

print("\nSaved:")
print("- heavy_usage_table_2024_25.csv (full table)")
print("- heavy_usage_pg_leaders_2024_25.csv (per game leaders)")
print("- heavy_usage_per36_leaders_2024_25.csv (per 36 leaders)")
print("- heavy_usage_share_leaders_2024_25.csv (team share leaders)")
print("- heavy_usage_pg.png")
print("- heavy_usage_per36.png")
print("- heavy_usage_share.png")

# ================================
# 7) Zip & Download Outputs (Colab)
# ================================
import os, zipfile
from google.colab import files

outputs = [
    "heavy_usage_table_2024_25.csv",
    "heavy_usage_pg_leaders_2024_25.csv",
    "heavy_usage_per36_leaders_2024_25.csv",
    "heavy_usage_share_leaders_2024_25.csv",
    "heavy_usage_pg.png",
    "heavy_usage_per36.png",
    "heavy_usage_share.png"
]

zip_name = "heavy_usage_outputs_2024_25.zip"

with zipfile.ZipFile(zip_name, "w") as z:
    for f in outputs:
        if os.path.exists(f):
            z.write(f)

files.download(zip_name)
```
